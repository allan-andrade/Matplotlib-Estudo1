## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercício 1

** Acompanhe estes passos: **
* ** Crie um objeto de figura chamado fig usando plt.figure () **
* ** Use add_axes para adicionar um eixo à tela de figura em [0,0,1,1]. Chame esse novo eixo de "ax". **
* ** Plote (x, y) nesses eixos e defina os rótulos e títulos para corresponder ao gráfico abaixo: **

```python
# Cria uma figura vazia
fig = plt.figure()

# Adiciona eixos à figura
ax = fig.add_axes([0, 0, 1, 1]) # esquerda, inferior, largura, altura (faixa de 0 a 1)

# Plota nos eixos especificados
ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Workshop Python')
```

## Exercício 3

** Crie o gráfico abaixo, adicionando dois eixos a um objeto de figura em [0,0,1,1] e [0,2,0,5, .4, .4] **

```python
# Cria uma figura vazia
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Eixo principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4]) # Eixos inseridos
```

** Agora use x, y e z arrays para recriar o gráfico abaixo. Observe os limites xlimits e y no gráfico inserido: **

```python
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1]) # Eixo principal
axes2 = fig.add_axes([.2,.5, .4, .4]) # Eixos inseridos

# Trabalhando no eixo maior
axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Z')
axes1.set_title('zoom')
axes1.set_ylim([0, 10000])
axes1.set_xlim([0, 105])

# Trabalhando no eixo menor
axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel('y')
axes2.set_title('zoom')
axes2.set_ylim([30, 50])
axes2.set_xlim([20, 22])
```

## Exercício 4

** Use plt.subplots (nrows = 1, ncols = 2) para criar o gráfico abaixo. **

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

** Agora trace (x, y) e (x, z) nos eixos. Note a largura de linha e o estilo objetivos. **

```python
fig, (axes1, axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y, color="blue", lw = 3, linestyle='--')
axes2.plot(x,z, "r")
plt.tight_layout()
```

** Veja se você pode redimensionar o gráfico adicionando o argumento figsize () em plt.subplots () apenas copiando e colando seu código anterior. **

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('título');
```

```python
fig,(axes1, axes2) = plt.subplots(nrows=1, ncols=2,figsize=(10,2))
axes1.plot(x,y, color="blue", lw = 3)
axes1.set_ylim([0, 200])
axes1.set_xlim([0, 100])
axes2.plot(x,z, "r", lw = 3, linestyle='--' )
plt.tight_layout()
```
